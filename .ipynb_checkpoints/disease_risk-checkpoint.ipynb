```python
# train.py
import pandas as pd
import numpy as np
import ast
import joblib
import os
import json
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from collections import Counter

# --- CONFIGURATION ---
TRAIN_PATH = 'dataset/22687585 (1)/release_train_patients/release_train_patients'
SYMPTOM_MAP_PATH = 'dataset/22687585 (1)/release_evidences.json'
OUTPUT_DIR = 'model_outputs/'
os.makedirs(OUTPUT_DIR, exist_ok=True)
N_SAMPLES_TO_USE = 100000  # Use a sample to keep library creation fast and memory-efficient

# --- MAIN SCRIPT ---
if __name__ == "__main__":
    print("=== MEDICAL RISK PREDICTION - TRAINING PHASE ===")
    
    # --- 1. Load and Downsample Data ---
    print(f"\n[1/5] Loading and downsampling data to {N_SAMPLES_TO_USE} samples...")
    df_full = pd.read_csv(TRAIN_PATH)
    df = df_full.sample(n=N_SAMPLES_TO_USE, random_state=42) if len(df_full) > N_SAMPLES_TO_USE else df_full
    
    with open(SYMPTOM_MAP_PATH, 'r', encoding='utf-8') as f:
        symptom_map_json = json.load(f)
    symptom_map = {k: v['name'].replace('_', ' ') for k, v in symptom_map_json.items() if isinstance(v, dict) and 'name' in v}
    print("✅ Data loaded.")
    print(f"   Total samples: {len(df)}")
    
    # Show original pathology distribution
    print("\nOriginal pathology distribution (top 15):")
    pathology_counts = df['PATHOLOGY'].value_counts().head(15)
    for pathology, count in pathology_counts.items():
        print(f"   {pathology}: {count}")
    
    # --- 2. Preprocess Data ---
    print("\n[2/5] Preprocessing data...")
    df['AGE'] = df['AGE'].fillna(df['AGE'].median())
    df['EVIDENCES_LIST'] = df['EVIDENCES'].apply(ast.literal_eval)
    df['symptom_text'] = df['EVIDENCES_LIST'].apply(lambda codes: " ".join([symptom_map.get(code, "") for code in codes]))
    
    # --- 3. Generate All Features ---
    print("\n[3/5] Generating features...")
    
    # Binary symptom features
    mlb = MultiLabelBinarizer()
    symptom_features = mlb.fit_transform(df['EVIDENCES_LIST'])
    df_symptoms = pd.DataFrame(symptom_features, columns=mlb.classes_, index=df.index)
    
    # Text-based features
    tfidf = TfidfVectorizer(max_features=5000, stop_words='english')
    text_features_tfidf = tfidf.fit_transform(df['symptom_text'])
    
    # Dimensionality reduction
    svd = TruncatedSVD(n_components=100, random_state=42)
    contextual_features = svd.fit_transform(text_features_tfidf)
    df_context = pd.DataFrame(contextual_features, index=df.index, columns=[f'context_{i}' for i in range(100)])
    
    # --- 4. Create Final Feature Library & IMPROVED RISK LABELS ---
    print("\n[4/5] Combining features and creating BALANCED risk categories...")
    
    # Categorical features
    sex_encoded = pd.get_dummies(df['SEX'], drop_first=True, prefix='SEX')
    
    # Combine all features
    X_library = pd.concat([df['AGE'], sex_encoded, df_symptoms, df_context], axis=1).fillna(0)
    
    # ============ IMPROVED RISK CLASSIFICATION ============
    print("Creating more balanced risk categories based on medical severity...")
    
    # High Risk - Life-threatening conditions requiring immediate intervention
    high_risk = [
        'Myocardial infarction',
        'Pulmonary embolism', 
        'Pneumonia',
        'Appendicitis',
        'Anaphylaxis',  # Actually very common in your data (3,646 cases)
        'Unstable angina',  # Common in your data (2,829 cases)
        'Panic attack',  # Can be severe/life-threatening if untreated (2,508 cases)
        'Acute rhinosinusitis',  # Can lead to serious complications if untreated
        'Bronchitis',  # Can progress to pneumonia if severe
        'Possible NSTEMI',  # Non-ST elevation myocardial infarction
        'Scombroid food poisoning',  # Can cause severe allergic-like reactions
        'Acute dystonic reaction',  # Medical emergency
        'Boerhaave syndrome',  # Life-threatening esophageal rupture
        'Chagas disease',  # Can be life-threatening if chronic
    ]
    
    # Moderate Risk - Serious conditions needing medical attention but not immediately life-threatening  
    moderate_risk = [
        'Malaria',
        'Dengue', 
        'Tuberculosis',
        'Gastroenteritis',
        'Anemia',  # Very common in your data (4,235 cases) - can be moderate to severe
        'HIV (initial infection)',  # Common in your data (2,914 cases)
        'GERD',  # Common in your data (2,733 cases) - can lead to complications
        'Localized edema',  # Common (2,778 cases) - can indicate serious underlying conditions
        'Cluster headache',  # Severe, debilitating condition
        'Atrial fibrillation',  # Heart rhythm disorder requiring monitoring
        'Acute laryngitis',  # Can compromise airway
        'Spontaneous pneumothorax',  # Lung collapse - needs immediate care
        'Guillain-Barré syndrome',  # Progressive paralysis
        'Myasthenia gravis',  # Muscle weakness disorder
        'Influenza',  # Can be severe, especially in vulnerable populations
        'Bronchospasm / acute asthma exacerbation',  # Respiratory emergency
        'Acute pulmonary edema',  # Heart/lung emergency
        'Pericarditis',  # Heart inflammation
        'Spontaneous rib fracture',  # Painful, may indicate underlying condition
    ]
    
    # Low Risk - Minor conditions, self-limiting, or routine care
    # URTI, Viral pharyngitis, and others will fall into this category by default
    
    # Apply the improved risk categorization
    y_library = df['PATHOLOGY'].apply(
        lambda pathology: 'High Risk' if pathology in high_risk 
        else 'Moderate Risk' if pathology in moderate_risk 
        else 'Low Risk'
    )
    
    # Display the new risk distribution
    risk_dist = y_library.value_counts()
    print("\n✅ NEW BALANCED RISK DISTRIBUTION:")
    for risk, count in risk_dist.items():
        percentage = (count / len(y_library)) * 100
        print(f"   {risk}: {count} ({percentage:.1f}%)")
    
    # Show which pathologies were classified as high/moderate risk
    print(f"\n📋 High Risk Pathologies Found in Data ({len([p for p in high_risk if p in df['PATHOLOGY'].values])}/{len(high_risk)}):")
    for pathology in high_risk:
        if pathology in df['PATHOLOGY'].values:
            count = (df['PATHOLOGY'] == pathology).sum()
            print(f"   ✓ {pathology}: {count} cases")
    
    print(f"\n📋 Moderate Risk Pathologies Found in Data ({len([p for p in moderate_risk if p in df['PATHOLOGY'].values])}/{len(moderate_risk)}):")
    for pathology in moderate_risk:
        if pathology in df['PATHOLOGY'].values:
            count = (df['PATHOLOGY'] == pathology).sum()
            print(f"   ✓ {pathology}: {count} cases")
    
    print(f"\n✅ Case library created with {X_library.shape[0]} patient vectors.")
    
    # --- 5. Save All Artifacts ---
    print("\n[5/5] Saving case library and all artifacts...")
    
    # Save the main artifacts
    np.save(os.path.join(OUTPUT_DIR, 'patient_library_vectors.npy'), X_library.values)
    np.save(os.path.join(OUTPUT_DIR, 'patient_library_labels.npy'), y_library.values)
    
    # Save preprocessing models
    joblib.dump(mlb, os.path.join(OUTPUT_DIR, 'symptom_encoder.pkl'))
    joblib.dump(tfidf, os.path.join(OUTPUT_DIR, 'tfidf_model.pkl'))
    joblib.dump(svd, os.path.join(OUTPUT_DIR, 'svd_model.pkl'))
    joblib.dump(symptom_map, os.path.join(OUTPUT_DIR, 'symptom_map.pkl'))
    joblib.dump(list(X_library.columns), os.path.join(OUTPUT_DIR, 'final_feature_columns.pkl'))
    
    # Save risk mapping for reference
    risk_mapping = {
        'high_risk': high_risk,
        'moderate_risk': moderate_risk,
        'distribution': dict(risk_dist)
    }
    joblib.dump(risk_mapping, os.path.join(OUTPUT_DIR, 'risk_mapping_info.pkl'))
    
    print(f"\n✅ All artifacts saved to '{OUTPUT_DIR}' directory.")
    print("✅ Training complete! The model now has BALANCED risk categories.")
    print(f"\n🎯 Expected improvement: With {risk_dist['High Risk'] + risk_dist['Moderate Risk']} high/moderate risk cases")
    print("   in training ({:.1f}%), predictions should be much more balanced!".format(
        ((risk_dist['High Risk'] + risk_dist['Moderate Risk']) / len(y_library)) * 100
    ))
    print("\n▶️  Next step: Run 'python predict.py' to see improved predictions!")
```

```python
# improved_predict.py - Memory-efficient with better high-risk detection
import pandas as pd
import numpy as np
import ast
import joblib
import os
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.utils.class_weight import compute_class_weight
from collections import Counter
import gc
from tqdm import tqdm

# --- IMPROVED CONFIGURATION ---
TEST_PATH = 'dataset/22687585 (1)/release_test_patients/release_test_patients'
ARTIFACTS_DIR = 'model_outputs/'
OUTPUT_FILE = os.path.join(ARTIFACTS_DIR, 'predictions.csv')

# Memory-efficient processing parameters
BATCH_SIZE = 1000  # Process test data in batches to save memory
K_NEIGHBORS = 20   # Slightly more neighbors for better voting

# High-risk detection parameters
HIGH_RISK_THRESHOLD = 0.3    # Lower threshold for high-risk classification
HIGH_RISK_BOOST = 2.0        # Strong boost for high-risk cases
USE_HYBRID_MODEL = True      # Use both similarity search AND ML classifier

class MemoryEfficientPredictor:
    def __init__(self, artifacts_dir):
        self.artifacts_dir = artifacts_dir
        self.load_artifacts()
        self.hybrid_classifier = None
        
    def load_artifacts(self):
        """Load all trained artifacts"""
        print("Loading artifacts...")
        self.X_library = np.load(os.path.join(self.artifacts_dir, 'patient_library_vectors.npy'), allow_pickle=True)
        self.y_library = np.load(os.path.join(self.artifacts_dir, 'patient_library_labels.npy'), allow_pickle=True)
        self.mlb = joblib.load(os.path.join(self.artifacts_dir, 'symptom_encoder.pkl'))
        self.tfidf = joblib.load(os.path.join(self.artifacts_dir, 'tfidf_model.pkl'))
        self.svd = joblib.load(os.path.join(self.artifacts_dir, 'svd_model.pkl'))
        self.feature_columns = joblib.load(os.path.join(self.artifacts_dir, 'final_feature_columns.pkl'))
        self.symptom_map = joblib.load(os.path.join(self.artifacts_dir, 'symptom_map.pkl'))
        
        print(f"Library loaded: {self.X_library.shape[0]} cases, {self.X_library.shape[1]} features")
        
        # Display risk distribution
        unique_risks, risk_counts = np.unique(self.y_library, return_counts=True)
        print("Library risk distribution:")
        for risk, count in zip(unique_risks, risk_counts):
            percentage = (count / len(self.y_library)) * 100
            print(f"  {risk}: {count} ({percentage:.1f}%)")
    
    def train_hybrid_classifier(self):
        """Train a Random Forest classifier to complement similarity search"""
        print("\nTraining hybrid ML classifier for high-risk detection...")
        
        # Use class weights to handle imbalance
        class_weights = compute_class_weight(
            'balanced', 
            classes=np.unique(self.y_library), 
            y=self.y_library
        )
        class_weight_dict = dict(zip(np.unique(self.y_library), class_weights))
        
        print(f"Class weights: {class_weight_dict}")
        
        # Split library data for training the classifier
        X_train, X_val, y_train, y_val = train_test_split(
            self.X_library, self.y_library, 
            test_size=0.2, random_state=42, 
            stratify=self.y_library
        )
        
        # Train Random Forest with balanced class weights
        self.hybrid_classifier = RandomForestClassifier(
            n_estimators=100,
            max_depth=15,
            min_samples_split=20,
            min_samples_leaf=10,
            class_weight=class_weight_dict,
            random_state=42,
            n_jobs=-1
        )
        
        # Scale features for better performance
        self.scaler = StandardScaler()
        X_train_scaled = self.scaler.fit_transform(X_train)
        
        self.hybrid_classifier.fit(X_train_scaled, y_train)
        
        # Evaluate on validation set
        X_val_scaled = self.scaler.transform(X_val)
        val_predictions = self.hybrid_classifier.predict(X_val_scaled)
        val_proba = self.hybrid_classifier.predict_proba(X_val_scaled)
        
        # Show validation performance
        val_dist = Counter(val_predictions)
        print(f"Validation predictions distribution:")
        for risk in ['High Risk', 'Moderate Risk', 'Low Risk']:
            count = val_dist.get(risk, 0)
            percentage = (count / len(val_predictions)) * 100
            print(f"  {risk}: {count} ({percentage:.1f}%)")
        
        return self.hybrid_classifier
    
    def preprocess_test_batch(self, df_batch):
        """Preprocess a batch of test data"""
        # Handle missing values
        df_batch['AGE'] = df_batch['AGE'].fillna(df_batch['AGE'].median())
        df_batch['EVIDENCES_LIST'] = df_batch['EVIDENCES'].apply(ast.literal_eval)
        df_batch['symptom_text'] = df_batch['EVIDENCES_LIST'].apply(
            lambda codes: " ".join([self.symptom_map.get(code, "") for code in codes])
        )
        
        # Generate features
        symptom_features = self.mlb.transform(df_batch['EVIDENCES_LIST'])
        df_symptoms = pd.DataFrame(symptom_features, columns=self.mlb.classes_, index=df_batch.index)
        
        text_features_tfidf = self.tfidf.transform(df_batch['symptom_text'])
        contextual_features = self.svd.transform(text_features_tfidf)
        df_context = pd.DataFrame(
            contextual_features, 
            index=df_batch.index, 
            columns=[f'context_{i}' for i in range(contextual_features.shape[1])]
        )
        
        # Encode categorical features
        sex_encoded = pd.get_dummies(df_batch['SEX'], drop_first=True, prefix='SEX')
        if 'SEX_M' not in sex_encoded.columns:
            sex_encoded['SEX_M'] = 0
        
        # Combine features and align with library schema
        X_batch = pd.concat([
            df_batch[['AGE']], 
            sex_encoded[['SEX_M']], 
            df_symptoms, 
            df_context
        ], axis=1)
        
        X_batch_aligned = X_batch.reindex(columns=self.feature_columns, fill_value=0)
        
        return X_batch_aligned
    
    def similarity_predict_batch(self, X_batch):
        """Memory-efficient similarity-based prediction for a batch"""
        batch_predictions = []
        batch_confidences = []
        
        # Calculate similarities in smaller chunks if needed
        similarities = cosine_similarity(X_batch.values, self.X_library)
        
        for i in range(similarities.shape[0]):
            # Get top K neighbors
            top_k_indices = np.argsort(similarities[i])[-K_NEIGHBORS:][::-1]
            top_k_labels = self.y_library[top_k_indices]
            top_k_scores = similarities[i][top_k_indices]
            
            # Risk-aware weighted voting with stronger high-risk bias
            vote_counts = Counter()
            for label, score in zip(top_k_labels, top_k_scores):
                if label == 'High Risk':
                    weight = HIGH_RISK_BOOST  # Strong boost for high risk
                elif label == 'Moderate Risk':
                    weight = 1.3
                else:
                    weight = 1.0
                
                vote_counts[label] += score * weight
            
            # Normalize votes
            total_weight = sum(vote_counts.values())
            if total_weight > 0:
                for label in vote_counts:
                    vote_counts[label] /= total_weight
            
            # Prediction with high-risk bias
            if vote_counts:
                # If high-risk has decent support, prefer it
                if 'High Risk' in vote_counts and vote_counts['High Risk'] >= HIGH_RISK_THRESHOLD:
                    prediction = 'High Risk'
                    confidence = vote_counts['High Risk']
                else:
                    prediction = vote_counts.most_common(1)[0][0]
                    confidence = vote_counts[prediction]
            else:
                prediction = 'Low Risk'
                confidence = 0.0
            
            batch_predictions.append(prediction)
            batch_confidences.append(confidence)
        
        return batch_predictions, batch_confidences
    
    def hybrid_predict_batch(self, X_batch):
        """Combine similarity search with ML classifier"""
        # Get similarity predictions
        sim_predictions, sim_confidences = self.similarity_predict_batch(X_batch)
        
        # Get ML classifier predictions
        X_scaled = self.scaler.transform(X_batch.values)
        ml_predictions = self.hybrid_classifier.predict(X_scaled)
        ml_probas = self.hybrid_classifier.predict_proba(X_scaled)
        
        # Get class labels for probability interpretation
        classes = self.hybrid_classifier.classes_
        
        # Combine predictions with bias toward high-risk
        final_predictions = []
        final_confidences = []
        
        for i in range(len(sim_predictions)):
            sim_pred = sim_predictions[i]
            ml_pred = ml_predictions[i]
            
            # Get ML confidence for the predicted class
            ml_class_idx = np.where(classes == ml_pred)[0][0]
            ml_confidence = ml_probas[i][ml_class_idx]
            
            # Decision logic: prioritize high-risk predictions
            if sim_pred == 'High Risk' or ml_pred == 'High Risk':
                final_pred = 'High Risk'
                # Use higher confidence between the two models
                final_conf = max(sim_confidences[i], ml_confidence)
            elif sim_pred == 'Moderate Risk' or ml_pred == 'Moderate Risk':
                final_pred = 'Moderate Risk'
                final_conf = max(sim_confidences[i], ml_confidence)
            else:
                # Both predict low risk
                final_pred = 'Low Risk'
                final_conf = (sim_confidences[i] + ml_confidence) / 2
            
            final_predictions.append(final_pred)
            final_confidences.append(final_conf)
        
        return final_predictions, final_confidences
    
    def predict_test_data(self, test_path):
        """Main prediction function with memory-efficient batch processing"""
        print(f"\nLoading test data from {test_path}...")
        df_test = pd.read_csv(test_path)
        print(f"Loaded {len(df_test)} test samples")
        
        # Train hybrid classifier if enabled
        if USE_HYBRID_MODEL:
            self.train_hybrid_classifier()
        
        # Process in batches to manage memory
        n_batches = (len(df_test) + BATCH_SIZE - 1) // BATCH_SIZE
        print(f"\nProcessing {n_batches} batches of size {BATCH_SIZE}...")
        
        all_predictions = []
        all_confidences = []
        all_patient_ids = []
        all_ages = []
        all_sexes = []
        all_num_symptoms = []
        
        for batch_idx in tqdm(range(n_batches), desc="Processing batches"):
            start_idx = batch_idx * BATCH_SIZE
            end_idx = min((batch_idx + 1) * BATCH_SIZE, len(df_test))
            
            batch_df = df_test.iloc[start_idx:end_idx].copy()
            
            # Preprocess batch
            X_batch = self.preprocess_test_batch(batch_df)
            
            # Predict batch
            if USE_HYBRID_MODEL and self.hybrid_classifier is not None:
                batch_preds, batch_confs = self.hybrid_predict_batch(X_batch)
            else:
                batch_preds, batch_confs = self.similarity_predict_batch(X_batch)
            
            # Store results
            all_predictions.extend(batch_preds)
            all_confidences.extend(batch_confs)
            all_patient_ids.extend(batch_df.index.tolist())
            all_ages.extend(batch_df['AGE'].tolist())
            all_sexes.extend(batch_df['SEX'].tolist())
            all_num_symptoms.extend([len(symptoms) for symptoms in batch_df['EVIDENCES_LIST']])
            
            # Clear memory
            del X_batch, batch_df
            gc.collect()
        
        # Create results DataFrame
        results_df = pd.DataFrame({
            'PATIENT_ID': all_patient_ids,
            'PREDICTED_RISK': all_predictions,
            'CONFIDENCE': all_confidences,
            'AGE': all_ages,
            'SEX': all_sexes,
            'NUM_SYMPTOMS': all_num_symptoms
        })
        
        return results_df

def main():
    print("=== IMPROVED MEDICAL RISK PREDICTION ===")
    print(f"Memory-efficient batch processing (batch size: {BATCH_SIZE})")
    print(f"High-risk detection optimized (threshold: {HIGH_RISK_THRESHOLD}, boost: {HIGH_RISK_BOOST})")
    print(f"Hybrid model enabled: {USE_HYBRID_MODEL}")
    
    # Initialize predictor
    predictor = MemoryEfficientPredictor(ARTIFACTS_DIR)
    
    # Make predictions
    results_df = predictor.predict_test_data(TEST_PATH)
    
    # Save results
    results_df.to_csv(OUTPUT_FILE, index=False)
    print(f"\n✅ Predictions saved to {OUTPUT_FILE}")
    
    # Display results summary
    print("\n=== PREDICTION SUMMARY ===")
    risk_distribution = results_df['PREDICTED_RISK'].value_counts()
    total_predictions = len(results_df)
    
    print("\nRisk Distribution:")
    for risk_level in ['High Risk', 'Moderate Risk', 'Low Risk']:
        count = risk_distribution.get(risk_level, 0)
        percentage = (count / total_predictions * 100) if total_predictions > 0 else 0
        print(f"  {risk_level}: {count} ({percentage:.1f}%)")
    
    print(f"\nConfidence Statistics:")
    print(f"  Average: {results_df['CONFIDENCE'].mean():.3f}")
    print(f"  Min: {results_df['CONFIDENCE'].min():.3f}")
    print(f"  Max: {results_df['CONFIDENCE'].max():.3f}")
    
    # High-risk cases summary
    high_risk_cases = results_df[results_df['PREDICTED_RISK'] == 'High Risk']
    if len(high_risk_cases) > 0:
        print(f"\n🚨 HIGH RISK CASES IDENTIFIED: {len(high_risk_cases)}")
        print("Sample high-risk predictions:")
        display_cols = ['PATIENT_ID', 'PREDICTED_RISK', 'CONFIDENCE', 'AGE', 'SEX', 'NUM_SYMPTOMS']
        print(high_risk_cases[display_cols].head(10).to_string(index=False))
        
        # Age distribution of high-risk cases
        print(f"\nHigh-risk age statistics:")
        print(f"  Average age: {high_risk_cases['AGE'].mean():.1f}")
        print(f"  Age range: {high_risk_cases['AGE'].min():.0f} - {high_risk_cases['AGE'].max():.0f}")
        
        # Save high-risk cases separately
        high_risk_file = os.path.join(ARTIFACTS_DIR, 'high_risk_cases.csv')
        high_risk_cases.to_csv(high_risk_file, index=False)
        print(f"  High-risk cases saved to: {high_risk_file}")
    
    print(f"\n✅ Improved prediction complete!")
    print(f"📊 Total processed: {total_predictions:,} patients")

if __name__ == "__main__":
    main()
```
